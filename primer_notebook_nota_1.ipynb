## Un Ejemplo de Pensamiento Algorítmico

Antes de mostrarte el código, pensemos en un problema sencillo de ingeniería: **Calcular el volumen de un reservorio cilíndrico de agua.**

La fórmula matemática es:


$$V = \pi \cdot r^2 \cdot h$$

Donde:

* $V$ es el volumen.
* $r$ es el radio de la base.
* $h$ es la altura del cilindro.

### El Código en Python

Aquí tienes una implementación básica. Observa que uso nombres de variables descriptivos, tal como haremos a lo largo del curso:

````python
# Importamos la librería matemática para obtener el valor preciso de PI
import math

def calcular_volumen_reservorio(radio_m, altura_m):
    """
    Calcula el volumen de un cilindro en metros cúbicos.
    """
    pi_valor = math.pi
    volumen_m3 = pi_valor * (radio_m ** 2) * altura_m
    return volumen_m3

# Datos de entrada (Imagina que estos vienen de un sensor o un formulario)
radio_input = 2.5  # metros
altura_input = 4.0 # metros

# Llamada a la función
resultado = calcular_volumen_reservorio(radio_input, altura_input)

print(f"El volumen del reservorio es: {resultado:.2f} m³")
````
