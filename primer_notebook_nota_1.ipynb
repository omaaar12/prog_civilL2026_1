# Problema de ingeniería (Nivel básico)

Un estudiante de **Ingeniería Civil o Agrícola** necesita calcular el **volumen de agua almacenado en un tanque rectangular**.

La fórmula del volumen es:

[
V = largo \times ancho \times altura
]

---

# 1. Definir Entradas, Proceso y Salidas

### Entradas

Datos que el usuario debe ingresar:

* largo del tanque (m)
* ancho del tanque (m)
* altura del tanque (m)

### Proceso

Operación matemática:

[
volumen = largo \times ancho \times altura
]

### Salida

Resultado que el programa mostrará:

* volumen del tanque en **metros cúbicos (m³)**

---

# 2. Algoritmo (explicado paso a paso)

1. Iniciar programa
2. Solicitar al usuario el **largo del tanque**
3. Solicitar el **ancho del tanque**
4. Solicitar la **altura del tanque**
5. Calcular el volumen usando la fórmula
6. Mostrar el resultado
7. Finalizar programa

---

# 3. Pseudocódigo

```
INICIO

leer largo
leer ancho
leer altura

volumen = largo * ancho * altura

mostrar volumen

FIN
```

---

# 4. Implementación en Python

```python
# Programa para calcular el volumen de un tanque rectangular
# Aplicación básica de variables y operadores

# Entrada de datos
largo = float(input("Ingrese el largo del tanque (m): "))
ancho = float(input("Ingrese el ancho del tanque (m): "))
altura = float(input("Ingrese la altura del tanque (m): "))

# Proceso
volumen = largo * ancho * altura

# Salida
print("El volumen del tanque es:", volumen, "m^3")
```

---

| Salida              | `print()`                  |
